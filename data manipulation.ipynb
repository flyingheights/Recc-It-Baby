```python
ds1 = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
ds2 = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
```

```python
ds1 = ds1.drop("movie_id", axis=1)
ds1
```

```python
ds = pd.merge(ds1, ds2)
ds = ds.drop(["id", "homepage", "popularity", "production_companies", "production_countries", "original_language", "budget", "revenue", "runtime", "spoken_languages", "status", "vote_count", "vote_average", "release_date"], axis = 1)
ds
```

```python
def extract_names(cell_value):
    try:
        data_list = json.loads(cell_value)
        return [entry["name"] for entry in data_list[:5]]
    except json.JSONDecodeError:
        return []
def extract_names1(cell_value):
    try:
        data_list = json.loads(cell_value)
        return [entry["name"] for entry in data_list]
    except json.JSONDecodeError:
        return []
ds['cast'] = ds['cast'].apply(extract_names)
ds['genres'] = ds['genres'].apply(extract_names1)
ds['keywords'] = ds['keywords'].apply(extract_names1)
ds
```

```python
'''def filter_directors(crew_str):
    try:
        crew_list = json.loads(crew_str)
        return [crew_dict.get('name') for crew_dict in crew_list if crew_dict.get('job') == 'Director']
    except json.JSONDecodeError:
        return []
ds['crew'] = ds['crew'].apply(filter_directors)'''
def extract_director_name(crew_str):
    try:
        crew_list = json.loads(crew_str)
        director_name = next((crew_dict.get('name') for crew_dict in crew_list if crew_dict.get('job') == 'Director'), None)
        return director_name
    except json.JSONDecodeError:
        return None
ds['crew'] = ds['crew'].apply(extract_director_name)
ds = ds.rename(columns={'crew': 'director'})
ds = ds.rename(columns={'original_title': 'title'})
ds
```

```python
ds.to_csv('new_dataset.csv', index = False)
```
